```python
from pyspark.ml.feature import VectorAssembler
feature_list = []
for col in df.columns:
    if col == '_c1024':
        continue
    elif col == '_c1025':
        continue
    else:
        feature_list.append(col)
assembler = VectorAssembler(inputCols=feature_list, outputCol="features")
```

```python
from pyspark.ml.feature import VectorAssembler
feature_list_test = []
for col in df_test.columns:
    if col == '_c1024':
        continue
    elif col == '_c1025':
        continue
    else:
        feature_list_test.append(col)
assembler_test = VectorAssembler(inputCols=feature_list_test, outputCol="features")
```

```python
df = assembler.transform(df)
```

```python
from pyspark.ml.classification import RandomForestClassifier

rf = RandomForestClassifier(featuresCol = 'features', labelCol = '_c1025', numTrees=10,maxDepth=10)
rfModel = rf.fit(train)
predictions = rfModel.transform(test)
```

```python

from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.sql.types import FloatType
preds_and_labels = predictions.select(['prediction','_c1025']).withColumn('label', F.col('_c1025').cast(FloatType())).orderBy('prediction')
```

```python
#select only prediction and label columns
preds_and_labels = preds_and_labels.select(['prediction','label'])

metrics = MulticlassMetrics(preds_and_labels.rdd.map(tuple))

print(metrics.confusionMatrix().toArray())
```

```python
array = metrics.confusionMatrix().toArray()
df_cm = pd.DataFrame(array, index = [i for i in "123456789"],
                  columns = [i for i in "123456789"])
plt.figure(figsize = (10,7))
sns.heatmap(df_cm, annot=True)
```

using MulticlassMetrics we can draw the confusion matrix. the more data exists in the diameter of the matrix the more accurate the model is. as we can see this model is as accurate as there are only a few mistakes in the predictions considering the large amout of data. the existence of these few numbers in places other than diameter indicates that our model makes some wrong predictions which could be due to data similarities in classes or our model could have difficulty classifying some data correctly.
